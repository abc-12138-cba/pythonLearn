```python
## 流式返回1
response = client.models.generate_content_stream(
    model="gemini-2.5-flash",
    contents=["人工智能是如何工作的"]
)
print(f"代码执行时间: {time.time() - start_time}秒")
num = 0
for chunk in response:
    num += 1
    print('~', num, chunk.text, end="")
```
